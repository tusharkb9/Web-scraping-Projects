```python
subreddit = reddit.subreddit('futurology')
```

```python
# Display the name of the subreddit
print('Display Name:', subreddit.display_name)

# Showing the title of the subreddit
print('Title:', subreddit.title)

# Showing the description of the subreddit
print('Description:', subreddit.description)
```

```python
# Scraping the top posts of all time
posts = subreddit.top('all')

post_dict = {'ID' : [], 'Title' : [], 'Score' : [], 'No. of comments' : [], 'Post URL' : []}

for post in posts:
    # ID of the posts
    post_dict['ID'].append(post.id)

    # Title of the posts
    post_dict['Title'].append(post.title)

    # Score of the posts
    post_dict['Score'].append(post.score)

    # No. of the comments
    post_dict['No. of comments'].append(post.num_comments)

    # URL of the posts
    post_dict['Post URL'].append(post.url)
```

```python
# Creating dataframe
futurology = pd.DataFrame(post_dict)

# Saving as csv
futurology.to_csv('futurology.csv', index = True)
```
